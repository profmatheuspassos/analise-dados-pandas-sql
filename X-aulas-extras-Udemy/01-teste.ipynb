```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xs = np.linspace(0, 9, 100)
ys = np.sin(xs)
plt.plot(xs, ys, label="Cool data")
plt.legend()
plt.show()
```

```python
xs = np.linspace(0, 9, 100)
ys = np.sin(xs)

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(xs, ys, label="Cool data")
ax.legend()

# Configurar limites do eixo X
ax.set_xlim(0, 9)

# Configurar valores de tick no eixo X
xticks = np.arange(0, 10)
ax.set_xticks(xticks)

plt.show()
```
