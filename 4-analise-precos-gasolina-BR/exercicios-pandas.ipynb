## Neste projeto, trabalharemos com dois arquivos csvs separados, contendo informações sobre o preço da gasolina no Brasil. 

### 1. Carregue os conjuntos de dados "gasolina_2000+.csv" e "gasolina_2010+.csv" em dois DataFrames diferentes e combine-os em um único DataFrame.

```python
dfGasolina2000 = pd.read_csv(arquivoGasolina2000, low_memory=False)
dfGasolina2010 = pd.read_csv(arquivoGasolina2010, low_memory=False)
dfCombinado = pd.concat([dfGasolina2000, dfGasolina2010], ignore_index=True)
```

### 4. Você deve ter percebido que as colunas DATA INICIAL e DATA FINAL estão formatadas como string. Utilizando o método pd.to_datetime(), converta ambas colunas para Timestamp / Datetime.

```python
dfCombinado[["DATA FINAL", "DATA INICIAL"]] = dfCombinado[["DATA FINAL", "DATA INICIAL"]].apply(lambda x: pd.to_datetime(x, format="%Y-%m-%d"))
```

### 9. Qual o preço médio de revenda da gasolina em maio de 2014 em São Paulo?

```python
# Filtrar por estado
dfSP = dfGasolinaComum[dfGasolinaComum["ESTADO"] == "SAO PAULO"]

# Filtrar por mês e ano usando a coluna DATA FINAL
dfMaio2014 = dfSP[(dfSP["DATA FINAL"].dt.month == 5) & (dfSP["DATA FINAL"].dt.year == 2014)]

# Calcular o preço médio de revenda
precoMedio = dfMaio2014["PREÇO MÉDIO REVENDA"].mean()

precoMedio
```
